```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# przygotowanie danych
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
def gauss(x, sigma=1, mu=0):
    """
    Funkcja gęstości prawdopodobieństwa dla rozkładu normalnego
    x - wartości, dla których obliczamy gęstość prawdopodobieństwa
    sigma - odchylenie standardowe
    mu - średnia
    """
    return stats.norm.pdf(x, mu, sigma)

dane_100= np.linspace(1, 100, 100)

mean_100 = np.round(np.mean(dane_100), 3)
median_100 = np.median(dane_100)
mode_100 = stats.mode(dane_100)

q0 = np.quantile(dane_100, 0)
q1 = np.quantile(dane_100, 0.25)
q2 = np.quantile(dane_100, 0.5) # mediana
q3 = np.quantile(dane_100, 0.75)
q4 = np.quantile(dane_100, 1)

plt.figure(figsize=(12, 8))
plt.plot(dane_100, gauss(dane_100, sigma=1, mu=3), label='100 próbek')
plt.axvline(mean_100, color='r', label='Średnia')
plt.axvline(median_100, color='g', label='Mediana')
plt.axvline(mode_100[0], color='orange', label='Moda')
plt.legend()
plt.show()
print(q0, q1, q2, q3, q4)
```

```python
plt.boxplot(dane_100)
plt.show()

```
